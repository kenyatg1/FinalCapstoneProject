```python
import requests
import json
import pandas as pd
import time
from keys import x_rapidapi_key, x_rapidapi_host
```

# API call

```python
def fetch_data(query):
    
    url = "https://zillow56.p.rapidapi.com/search"

    # query header parameters
    headers = {
	"x-rapidapi-key": x_rapidapi_key,
	"x-rapidapi-host": x_rapidapi_host
}
    response = requests.get(url, headers=headers, params=query).json()
    time.sleep(1)

    return pd.json_normalize(response['results'])
```

```python
# sets counter for ['page'] in each query
page = "0"
repeat = 30

# empty dataframe to append to
data = pd.DataFrame()

# query parameters
querystring = [{"location":"new york, ny", "page": page , "output":"json","status":"forSale","sortSelection":"priorityscore","listing_type":"by_agent",
                "isSingleFamily":"true","doz":"any"},
               {"location":"miami, fl", "page": page ,"output":"json","status":"forSale","sortSelection":"priorityscore","listing_type":"by_agent",
                "isSingleFamily":"true","doz":"any"},
               {"location":"charlotte, nc", "page": page ,"output":"json","status":"forSale","sortSelection":"priorityscore","listing_type":"by_agent",
                "isSingleFamily":"true","doz":"any"}
              ]

# requests 3 pages of properties from each city
for city in querystring:
    for counter in range(repeat):
        r = fetch_data(city)
        print(r)
        data = pd.concat([data, r], ignore_index=True)
        # increments pages for each city
        city['page'] = int(city['page']) + 1
        city['page'] = str(city['page'])  
        
```

# Export Data

```python
data.to_csv('API.csv')
```
