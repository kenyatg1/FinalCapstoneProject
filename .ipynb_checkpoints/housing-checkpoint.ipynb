# Fix upper limit

```python
# Any price above 750000 gets excluded
df = df.loc[df['price']<=750000]
```

```python
df.info()
```

### Change suburbs to proper city names

```python
df['city'].unique()
```

```python
# get zip codes by city
zipcodes = df['zipcode'].unique()
zipcodes.sort()
zipcodes
```

```python
# set lists for zip codes of each city
new_york_zip = []
charlotte_zip  = []
miami_zip = []

for zip in zipcodes:
    if zip < 20000:
        new_york_zip.append(zip)
    elif 20000 < zip < 30000:
        charlotte_zip.append(zip)
    else:
        miami_zip.append(zip)
```

```python
# change city based off zip codes
df.loc[df['zipcode'].isin(new_york_zip), 'city'] = 'New York'
df.loc[df['zipcode'].isin(miami_zip), 'city'] = 'Miami'
df.loc[df['zipcode'].isin(charlotte_zip), 'city'] = 'Charlotte'
```

```python
# check if it worked
df['city'].unique()
```
